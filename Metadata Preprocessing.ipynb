## Preprocessing meta properties

```python
prop = df_meta['properties'].tolist()
iid = df_meta['item_id'].tolist()

prop_l = {}
split = []
for i in range(len(prop)):
    temp = prop[i].split("|")
    prop_l[iid[i]] = temp
    split.append(temp)
```

```python
df_meta['split']=split
```

```python
df_meta.head(2)
```

```python
prop_t = {}

for p in split:
    for q in p:
        if q in prop_t.keys():
            prop_t[q] += 1
        else:
            prop_t[q] = 1
```

```python
# convert properties to vectors
prop_v = []

for value in split:
    temp = []
    for p in prop_t.keys():
        if p in value:
            temp.append(1)
        else:
            temp.append(0)
    prop_v.append(temp)
df_meta['vector']=prop_v
```

```python
# add session as a properties
ref_tr = df_tr['reference'].tolist()
session = df_tr['session_id'].tolist()
ref_se = {} # reference is the key and session id is the value
for n in range(len(ref_tr)):
    if ref_tr[n] in ref_se.keys():
        temp = ref_se[ref_tr[n]]
        if session[n] in temp:
            continue
        else:
            temp.append(session[n])
        ref_se[ref_tr[n]]=temp
    else:
        ref_se[ref_tr[n]] = [session[n]]
```
